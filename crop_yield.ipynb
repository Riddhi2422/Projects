```python
df=df.drop('Yield (tons/ha)',axis=1)
df
```

```python
df=df[df['Crop Year']>=2004]
df
```

```python
df=df.join(pd.get_dummies(df['Season']))
df=df.join(pd.get_dummies(df['Crop Name']))
df=df.join(pd.get_dummies(df['State Name']))
df=df.join(pd.get_dummies(df['Soil Type']))
df
```

```python
df['Yield (tons/ha)']=df['Production (tons)']/ df['Area (ha)']
df
```

```python
df=df.drop('Production (tons)' , axis=1)
```

```python
df=df.drop('District Name' , axis=1)
df=df.drop('Season' , axis=1)
df=df.drop('Crop Name' , axis=1)
df=df.drop('State Name' , axis=1)
df=df.drop('Soil Type' , axis=1)
df
```

```python
from sklearn import preprocessing
x=df[['Area (ha)']].values.astype(float)
x

min_max_scaler=preprocessing.MinMaxScaler()

x_scaled= min_max_scaler.fit_transform(x)
x_scaled

df['Area (ha)']=x_scaled
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop(columns='Yield (tons/ha)')
y=df['Yield (tons/ha)']
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.3 , random_state= 42)
```

```python

import numpy as np
import matplotlib.pyplot as plt
import seaborn as seabornInstance
from sklearn.linear_model import LinearRegression
from sklearn import metrics
%matplotlib inline
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train= sc.fit_transform(x_train)
x_test= sc.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regr=RandomForestRegressor(max_depth=2 ,random_state=0 , n_estimators=100)
regr.fit(x_train, y_train)
y_pred= regr.predict(x_test)


```

```python
from sklearn.model_selection import cross_val_score
cv_scores = cross_val_score(regr, x_train, y_train, cv=5)
cv_scores
```

```python
from sklearn.metrics import mean_squared_error as mse
from sklearn.metrics import mean_absolute_error as mae
from sklearn.metrics import r2_score

print('MSE=' ,mse(y_pred, y_test))
print('MAE=', mae(y_pred, y_test))
print('R2_score=' ,r2_score(y_pred, y_test))
```
